## Python environment setup

The GUI should be run in an adequate python environment (with recent versions of the panel and bokeh libraries in particular). One such environment has been prepared for the fit_ic community. To use it, you first need to activate it in your own jupyter-hub account:
1. Open a terminal inside the jupyter hub (**File > New > Terminal**)
2. Then copy paste and run the command `~/project/fit_ic/.fitic/envs/fitic/bin/python -m ipykernel install --name FITIC --user`
3. Restart your jupyter-hub server: click on **File > Hub Control Panel**, then click on **Stop my server**, wait for 30 seconds, and finally click on **Start my server**
4. Come back to this notebook, and make sure that the "FIT-IC" kernel is selected (click on the button on the top-right corner, which shows "Python 3 (ipykernel)", by default, and switch to "FIT-IC"
